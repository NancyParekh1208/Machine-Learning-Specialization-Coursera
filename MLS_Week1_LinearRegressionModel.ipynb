```python
import numpy as np
from matplotlib import pyplot as plt

x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])

print(f"x_train = {x_train}, y_train = {y_train}")

print(f"Number of training examples in x_train: {x_train.shape[0]}")

w = 200
b = 100

def compute_housing_prices(x_train, w, b,):
  m = x_train.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x_train[i] + b
  return f_wb

f_wb = compute_housing_prices(x_train,w, b)

# Plotting the model prediction of house prices

plt.plot(x_train , f_wb , c = 'r', label = "Our Prediction")

# Plotting the data points
plt.scatter(x_train, y_train, marker = 'x', c = 'b', label = "Actual Values")

# Setting title and axis label's
plt.xlabel('Size of plot (1000 sq ft)')
plt.ylabel('Price of the plot (Price in 1000s of dollars)')
plt.title('Housing Prices')
plt.legend()
plt.show()
```
